```python
# Function to reduce the memory usage of a DataFrame.
def reduce_memory(df):
    for col in df.columns:
        if df[col].dtype == 'float64':
            df[col] = df[col].astype('float32')
        if df[col].dtype == 'int64':
            df[col] = df[col].astype('int32')
    return df

# Generator function to load data in chunks.
def data_generator(df, chunksize=10000):
    for i in range(0, df.shape[0], chunksize):
        yield df.iloc[i:i+chunksize]
```

```python
df = reduce_memory(pd.read_csv("data/Dataset.csv"))
```

```python
from scipy.sparse import coo_matrix
from sklearn.neighbors import NearestNeighbors

# map each user and item to a unique numeric value
user_ids = df['UserID'].astype('category').cat.codes
item_ids = df['GameID'].astype('category').cat.codes

# Get the unique user and game ids
unique_user_ids = df['UserID'].astype('category').cat.categories
unique_item_ids = df['GameID'].astype('category').cat.categories

# create a sparse matrix
user_game_matrix = coo_matrix((df['Heure_jouee'], (user_ids, item_ids)))

# Fit the model
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(user_game_matrix)

# Get top 5 recommendations for first user
distances, indices = model_knn.kneighbors(user_game_matrix.getrow(0), n_neighbors=6)

# Get the games played by the first user
games_played_by_user = df[df['UserID'] == unique_user_ids[0]]['GameID'].values

# Get the recommended users excluding the first user
recommended_users = [unique_user_ids[i] for i in indices.flatten()[1:]]

# Find games played by the recommended users but not played by the first user
games_played_by_recommended_users = df[df['UserID'].isin(recommended_users)]['GameID'].values
recommended_games = set(games_played_by_recommended_users) - set(games_played_by_user)

# Sort recommended games by total playtime in descending order
sorted_recommended_games = sorted(recommended_games, key=lambda game_id: df[df['GameID'] == game_id]['Heure_jouee'].sum(), reverse=True)

# Print the top 5 games recommended for the first user based on total playtime of recommended users
print(f'Top 5 games recommended for the first user based on total playtime of recommended users:')
for game_id in sorted_recommended_games[:5]:
    game_name = df[df['GameID'] == game_id]['Game'].values[0]
    print(game_name)
```
